```python
import pandas as pd
import numpy as np
integrated_data = pd.read_parquet('data/integrated_data.parquet')
integrated_data
```

```python
# Calculate the percentage of missing values for each column
missing_percentage = integrated_data.isnull().mean() * 100

# Filter columns where the percentage of missing values is greater than 30%
columns_with_high_missing_percentage = missing_percentage[missing_percentage > 10].index

# Select the columns with high missing percentage from the original_data DataFrame
columns_filtered = integrated_data[columns_with_high_missing_percentage]

# Print the columns with high missing percentage
print(columns_filtered.columns)
```

```python
data_cleaned = integrated_data.drop(columns_with_high_missing_percentage, axis=1)

# Create a new column "Категориальная переменная" based on the "Дата иска" column
data_cleaned['Категориальная переменная'] = np.where(data_cleaned['Статус'] == 'Действующая', False, True)

data_cleaned
```

```python
data_cleaned = data_cleaned.dropna()
```

```python
# Save the data_cleaned DataFrame
data_cleaned.to_excel('data/data_cleaned.xlsx', index=False)
data_cleaned.to_parquet('data/data_cleaned.parquet', index=False)
```
